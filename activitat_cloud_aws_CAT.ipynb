### 1.1. Introducció del projecte

El projecte consisteix en dissenyar i implementar una pipeline de dades a AWS per processar les vendes diàries d’una cadena de botigues de roba repartides per diferents ciutats de Catalunya. Les dades utilitzades són sintètiques i han estat generades específicament per simular un escenari realista de vendes per botiga i dia.

Cada botiga genera un fitxer CSV diari amb les vendes realitzades durant la jornada. Cada fitxer conté informació de les vendes realitzades en una botiga concreta, com l’identificador de la botiga, la data, l’hora de la venda, el producte venut, el departament, la talla, el preu, les unitats venudes, el color i la descripció del producte.

L’objectiu del projecte és emmagatzemar aquests fitxers a AWS, validar-ne l’estructura, processar les dades i carregar-les en una base de dades relacional perquè quedin preparades per a l’anàlisi. Com a resultat final, les transaccions quedaran emmagatzemades en una base de dades Amazon Aurora compatible amb MySQL i en S3. Com a validació addicional, Amazon Athena permetrà consultar amb SQL els fitxers emmagatzemats a S3 i comparar-ne els resultats amb les dades carregades a la base de dades.

La pipeline s’inicia quan es puja un fitxer CSV a la carpeta `entrada/` del bucket S3. Aquesta pujada activa una funció Lambda que valida el fitxer i el mou a `validat/` o `error/`. Quan un fitxer queda a `validat/`, S3 envia un missatge a una cua SQS, que activa una segona funció Lambda encarregada de processar el fitxer i carregar les dades a RDS.

El flux principal cobreix el procés ETL (Extracció, Transformació i Càrrega) de les dades. La pipeline és escalable i permet processar múltiples fitxers de manera concurrent, gràcies a la combinació de S3, Lambda i SQS.

### 1.2. Tipus de dades utilitzades

Les dades utilitzades representen vendes de productes de roba. Per una banda, es disposa d’un fitxer `productes.csv`, que conté el catàleg de productes disponibles. Per altra banda, cada botiga genera un fitxer CSV de vendes amb les operacions realitzades durant un dia concret.

El fitxer `productes.csv` té els següents camps:

```text
id_producte
departament
descripcio
preu
```

El camp `id_producte` és l’identificador únic del producte i correspon a una cadena alfanumèrica. El camp `departament` indica la categoria del producte, com home, dona, nen o nena. El camp `descripcio` conté una breu descripció del producte i el camp `preu` indica el preu unitari.

Cadascun dels fitxers de vendes té els següents camps:

```text
id_botiga
data
hora
id_producte
departament
talla
preu
unitats
color
descripcio
```

Els serveis seleccionats són:

### 1.4. Justificació dels serveis AWS

#### 1.4.1 Amazon S3

Amazon S3 s’utilitza com a sistema principal d’emmagatzematge de fitxers. Dins del bucket es creen diferents carpetes lògiques per separar les fases del procés:

```text
entrada/
validat/
error/
processat/
```

La carpeta `entrada/` conté els fitxers CSV originals pujats al sistema. La carpeta `validat/` conté els fitxers que han superat la validació inicial feta amb Lambda. La carpeta `error/` conté els fitxers rebutjats o que no compleixen l’estructura esperada. Finalment, la carpeta `processat/` conté els fitxers que ja han estat processats. Com a proposta futura, els fitxers de `processat/` es conservarien durant dos anys i després es traslladarien a S3 Glacier mitjançant una política de cicle de vida.

S3 és adequat perquè és un servei gestionat, escalable i molt utilitzat com a repositori de dades. També permet mantenir la traçabilitat del procés, ja que es poden conservar separadament els fitxers originals, els validats, els erronis i els processats.

#### 1.4.2 AWS Lambda

AWS Lambda s’utilitza en dues parts diferents de la pipeline.

La primera funció Lambda s’encarrega de validar els fitxers CSV que arriben a la carpeta `entrada/` del bucket S3. Aquesta funció llegeix el fitxer CSV, comprova que la capçalera tingui l’estructura esperada i valida els registres abans de permetre que continuïn dins de la pipeline. Si el fitxer és correcte, el mou a `validat/`; si detecta algun error, el mou a `error/`.

La segona funció Lambda s’encarrega del processament posterior. Aquesta funció s’activa a partir dels missatges de SQS, llegeix el fitxer validat, processa el contingut del CSV i carrega les dades a la base de dades RDS.

Lambda és adequada perquè és un servei serverless basat en esdeveniments. Permet executar codi només quan cal, sense mantenir una màquina encesa permanentment per validar o processar fitxers.

#### 1.4.3 Amazon CloudWatch

Amazon CloudWatch s’utilitza per consultar els logs generats per les funcions Lambda durant la validació i el processament dels fitxers. Això permet veure quan s’ha executat cada funció, quin fitxer s'ha processat, si el procés ha estat correcte o si s’ha produït algun error.

#### 1.4.4 Amazon SQS

Amazon SQS s’utilitza per desacoblar la fase de validació de la fase de processament. Quan un fitxer és validat i queda a la carpeta `validat/`, S3 envia un missatge a una cua SQS amb la informació del fitxer.

Aquesta cua permet que la segona funció Lambda processi els fitxers validats de manera desacoblada i controlada. D’aquesta manera, la validació i la càrrega a RDS no depenen directament l’una de l’altra, i la pipeline és més robusta davant errors puntuals.

SQS també facilita el control dels fitxers pendents de processar. Quan la Lambda de processament acaba correctament, SQS elimina el missatge associat de la cua.

#### 1.4.5 Amazon EC2

Amazon EC2 s’utilitza per a tasques d’administració i proves. Des d’aquesta instància es pot connectar a RDS, crear les taules de la base de dades, carregar el fitxer `productes.csv` i executar consultes SQL per verificar que les dades s’han carregat correctament. EC2 també s’utilitza per generar la capa de Lambda amb les dependències necessàries per connectar-se a RDS.

EC2 també serveix com a entorn de suport per preparar fitxers, executar comandes de prova i validar la connexió amb la base de dades. No és el component principal de processament de la pipeline, ja que aquesta part queda automatitzada amb Lambda.

#### 1.4.6 Amazon RDS

Amazon RDS s’utilitza com a base de dades relacional final. En aquest projecte s’utilitza MySQL. Les dades netes generades durant el processament es carreguen en dues taules principals:

```text
productes
vendes
```

La taula `productes` conté la informació dels productes, com l’identificador, el departament, la descripció i el preu. La taula `vendes` conté les vendes processades, com la botiga, el producte, la data, l’hora i la quantitat venuda. La relació entre les dues taules es fa mitjançant el camp `id_producte`.

RDS permet fer consultes SQL per obtenir indicadors com el nombre total de vendes, les unitats venudes, les vendes per botiga, les vendes per producte o les vendes per departament combinant la informació de `vendes` i `productes`.

RDS és adequat perquè les dades finals són estructurades i tabulars, i perquè MySQL permet consultar-les de manera senzilla i coneguda.

#### 1.4.7 Amazon Athena

Amazon Athena s’utilitza com a servei addicional d’anàlisi. No substitueix RDS, sinó que serveix per validar i fer consultes exploratòries directament sobre els fitxers processats de S3.

Això permet comprovar que el fitxer processat de S3 és coherent amb les dades carregades a RDS. Per exemple, es pot comparar el nombre de registres consultat amb Athena sobre `processat/` amb el nombre de files carregades a la taula `vendes` de RDS.

Athena s’ha triat perquè és serverless, fàcil de provar, s’integra bé amb S3 i permet afegir una capa d’anàlisi sense complicar l’arquitectura.

#### 1.4.8 Amazon VPC

La part de xarxa es gestiona amb Amazon VP i es distribueix en dues zones de disponibilitat i inclou subxarxes públiques i privades. RDS i EC2 es despleguen dins la mateixa VPC, en una configuració que permeti la comunicació entre ells. També és necessari que la Lambda de processament pugui connectar-se a RDS per carregar-hi les dades.

La idea és que RDS no estigui obert innecessàriament a internet i que només els components autoritzats puguin accedir-hi. Com que s’utilitza MySQL, RDS escolta habitualment pel port 3306.

A més, utilitzem un VPC Endpoint per a S3 per evitar que les comunicacions entre Lambda i S3 passin per internet públic. Això millora la seguretat i manté el trànsit dins la xarxa d’AWS.

### 1.5. Processament de dades i integració entre serveis

El processament de dades segueix un esquema ETL: extracció, transformació i càrrega.

A nivell d’integració, el flux és el següent: els fitxers CSV es carreguen a S3 `entrada/`; una primera Lambda valida el fitxer i el mou a `validat/` o `error/`; quan el fitxer queda a `validat/`, S3 envia a SQS un missatge amb la informació necessària perquè la Lambda de processament recuperi el fitxer des de `validat/`; la segona Lambda consumeix el missatge, processa el CSV i carrega les dades a RDS MySQL; un cop finalitza correctament el processament, el fitxer es mou a S3 `processat/`. CloudWatch registra els logs de les funcions Lambda i Athena permet consultar els fitxers processats guardats a S3.


#### 1.5.1 Extracció

La fase d’extracció comença quan els fitxers CSV de vendes arriben a Amazon S3, dins la carpeta `entrada/`.

Cada fitxer representa les vendes d’una botiga en una data concreta. Aquesta fase no modifica les dades, sinó que les deixa disponibles en el repositori inicial.

#### 1.5.2 Validació inicial

Quan es puja un fitxer CSV a S3 `entrada/`, la primera funció Lambda s’activa mitjançant un esdeveniment de S3. La funció llegeix el fitxer, comprova que la capçalera tingui l’estructura esperada i valida els registres abans de permetre que continuïn dins de la pipeline.

Si el fitxer supera la validació, es mou a `validat/`; si es detecta algun error, es mou a `error/`. El resultat de l’execució i els possibles errors queden registrats a CloudWatch.

Quan un fitxer arriba a `validat/`, S3 envia a SQS un missatge amb la informació necessària perquè la Lambda de processament recuperi el fitxer des de `validat/`.

#### 1.5.3 Processament

La transformació i càrrega principal es fa amb una segona funció Lambda, activada pels missatges de SQS. Aquesta funció consumeix el missatge, recupera el fitxer validat des de S3 i interpreta el contingut del CSV.

Les operacions principals consisteixen a convertir els camps `data`, `hora` i `unitats` als tipus adequats i preparar la inserció de les dades a la taula `vendes`. Durant la càrrega, el camp `unitats` del CSV s’insereix al camp `quantitat` de la taula `vendes`.

La relació amb la taula `productes` es controla mitjançant el camp `id_producte` i la clau forana definida a RDS, que impedeix inserir vendes associades a productes inexistents.

#### 1.5.4 Càrrega

La Lambda de processament carrega les dades a RDS compatible amb MySQL. Quan la càrrega finalitza correctament, mou el fitxer des de S3 `validat/` fins a `processat/`.

La càrrega a RDS permet conservar les dades en format relacional i consultar-les posteriorment amb SQL. Després d’una execució correcta de la Lambda, SQS elimina automàticament el missatge associat de la cua.

No s’implementa en aquesta pràctica, però es preveu conservar els fitxers de `processat/` durant dos anys abans d’arxivar-los a S3 Glacier.

#### 1.5.5 Consulta i verificació a RDS

Un cop les dades estan carregades a RDS, es poden executar consultes SQL des de la instància EC2 per comprovar que la càrrega s’ha fet correctament i obtenir indicadors bàsics.

Per exemple:

```sql
SELECT COUNT(*) AS total_vendes
FROM vendes;
```

```sql
SELECT *
FROM vendes
ORDER BY data DESC, hora DESC
LIMIT 5;
```

Aquestes consultes permeten verificar el nombre de registres inserits, revisar les vendes més recents i comprovar que les dades queden disponibles per a l’anàlisi posterior.

#### 1.5.6 Validació addicional amb Athena

Athena consulta directament els fitxers de S3 `processat/`. Per fer-ho, es crea una taula externa que apunta a aquesta ubicació i interpreta els CSV processats com una única taula consultable amb SQL.

Per exemple, es pot executar una consulta de recompte:

```sql
SELECT COUNT(*) AS total_vendes_processades
FROM vendes_processades;
```

També es poden fer consultes agregades, com el total d’unitats venudes per producte o l’import total de les vendes:

```sql
SELECT id_producte, SUM(unitats) AS total_unitats
FROM vendes_processades
GROUP BY id_producte
ORDER BY total_unitats DESC;
```

```sql
SELECT SUM(preu * unitats) AS import_total
FROM vendes_processades;
```

### Validació de fitxers CSV d'entrada (S3, Lambda, CloudWatch, IAM)

En aquesta part es configura el bucket de S3 i la funció Lambda que valida els fitxers d’entrada. La funció Lambda s’encarrega de validar els fitxers CSV que es pugen a la carpeta `entrada/` del bucket de S3, i mou els fitxers vàlids a la carpeta `validat/` i els fitxers amb errors a la carpeta `error/`. A més, es comprova que el desencadenador de S3 està funcionant correctament i que els logs de CloudWatch mostren la informació esperada.

1. Crear un bucket a S3 per a emmagatzemar els fitxers d'entrada, els fitxers validats i els fitxers amb errors.
   - Obrir **AWS Management Console** i busca el servei **S3** des del cercador de serveis.
   - Si no s'ha creat cap bucket anteriorment, clicar al botó **Create Bucket**. Altrament, entrar a **Buckets** i crear el bucket del projecte, amb un nom únic (`amzn-s3-uoc-2026-bucket`) i la regió del laboratori.
   - Crear les carpetes `entrada/`, `validat/` i `error/` a l'interior del bucket mitjançant el botó **Crear carpeta**.
2. En AWS Academy, no es poden crear rols personalitzats, utilitzem el rol `LabRole`.  En un entorn AWS normal, cal crear un rol d'execució per a Lambda amb el nom `CsvValidationRole` i assignar-li les polítiques gestionades `AWSLambdaBasicExecutionRole` i `AmazonS3FullAccess`:
   - Anar al cercador de serveis i obrir **IAM**.
   - En el plafó lateral de l'esquerra, clicar a **Roles** i després a **Crear rol**.
   - En la primera pantalla, seleccionar **Servicio de  AWS service** i després **Lambda** com a cas d'ús.
   - En la següent pantalla, buscar i seleccionar les polítiques gestionades `AWSLambdaBasicExecutionRole` i `AmazonS3FullAccess`.
   - En l'última pantalla de **Asignar nombre, revisar y crear**, afegir l'etiqueta `Name` amb el valor `Rol validació fitxers CSV`. Clicar a **Crear rol**.
3. Crear la funció Lambda que s'encarregarà de validar els fitxers CSV d'entrada:
   - Anar al cercador de serveis i obrir **Lambda**.
   - Crear la funció amb **Crear función** i seleccionar **Crear desde cero**, proporcionar un nom (`ValidarFitxersCsv`), triar python com entorn d'execució.
   - Desplegar la configuració personalitzada, habilitar **Rol de ejecución personalizada** i seleccionar el rol d'execució `LabRole` (en un entorn AWS normal utilitzaríem el rol creat amb IAM).
   - A la pestanya **Configuration**, anar a **Agregar desencadenadores**:
     - Triar **S3** com a desencadenador.
     - Seleccionar el bucket creat anteriorment i el prefix `entrada/`.
     - Deixar les opcions per defecte i clicar a **Agregar**.
4. Provar la funció Lambda i el desencadenador de S3:
   - Des de S3 pugem un fitxer CSV de prova a `entrada/` mitjançant el botó **Cargar**.
   - Des de Lambda, anar a la pestanya **Monitorear**, clicar al botó **Ver registros de CloudWatch** a **CloudWatch Logs**.
   - Des de CloudWatch, revisar els logs de la funció Lambda per veure que el fitxer s’ha validat i que s’ha mogut a `validat/` o `error/` segons correspongui.
   - Des de S3, verificar al bucket que els CSV vàlids queden a `validat/` i els erronis a `error/`.

Fins aquest punt, les crides de la funció lambda al servei S3 es fan a través d'internet públic. Per a millorar la seguretat, es pot configurar un endpoint de VPC per a S3, que permet que les comunicacions entre Lambda i S3 es facin a través de la xarxa privada d'AWS en lloc d'internet públic. Això es fa des de la consola de VPC, creant un endpoint per a S3 i associant-lo amb la VPC on està desplegada la funció Lambda. Un cop configurat l'endpoint, les crides de Lambda a S3 es redirigiran automàticament a través de l'endpoint, millorant la seguretat i el rendiment de les comunicacions entre els serveis.

Posteriorment configurarem la replicació de la base de dades RDS amb dos zones de disponibilitat, per aquest motiu, cal configurar la VPC amb dues zones de disponibilitat i associar la funció Lambda a aquesta VPC per a que les comunicacions entre Lambda i S3 es facin a través de la xarxa privada d'AWS.

   - Selecciona 2 zones de disponibilitat
   - Seleccionar 2 subxarxes públiques (per exemple, `10.0.1.0/24` i `10.0.3.0/24`).
   - Seleccionar 2 subxarxes privades (per exemple, `10.0.2.0/24` i `10.0.4.0/24`).

      - **Bloque de CIDR de la subred pública 2 en us-east-1b**: `10.0.3.0/24`
      - **Bloque de CIDR de la subred privada 1 en us-east-1a**: `10.0.2.0/24`
      - **Bloque de CIDR de la subred privada 2 en us-east-1b**: `10.0.4.0/24`
   - Seleccionar la porta d'enllaç **Zonal** i seleccionar **En 1 AZ**.
   - **Puntos de enlace de la VPC** marcar la casella **Habilitar el punto de enlace de la VPC para S3**.
   - Clicar a **Crear VPC**.

### Carregar fitxers CSV validats a base de dades (S3, Lambda, CloudWatch, IAM, RDS, EC2, SQS)

En aquesta part s'utilitza una cua de SQS per a gestionar els fitxers CSV que han estat validats per carregar-los a una base de dades RDS. Els missatges de la cua s'utilitzen per a coordinar el processament i la inserció de dades. Quan un fitxer és validat i mogut a `validat/`, S3 envia un missatge a la cua de SQS amb la informació del fitxer. Una funció Lambda està configurada per a llegir els missatges de la cua, carregar les dades del fitxer CSV a una base de dades RDS i eliminar el missatge de la cua un cop processat. A més, es comprova que el desencadenador de SQS està funcionant correctament i que els logs de CloudWatch mostren la informació esperada.

1. Crear una cua de SQS per a gestionar els fitxers CSV validats:
   - Anar al cercador de serveis i obrir **SQS**.
   - Clicar a **Crear cola** i seleccionar **Cola estàndard**.
   - Proporcionar un nom per a la cua (`validatedCsvQueue`) i deixar les opcions per defecte. Clicar a **Crear cola**.

2. Per afegir els permisos necessaris a S3 per a publicar missatges a la cua de SQS, cal modificar la política d'accés de la cua de SQS per a permetre que S3 publiqui missatges. En un entorn AWS normal, es faria així:
   - Anar a la cua de SQS creada anteriorment i clicar a la pestanya **Permisos**.
   - Clicar a **Editar política de acceso**
   - Clicar a **Generador de políticas**
     - En el pas 1, seleccionar el tipus de política **SQS Queue Policy**.
     - En el pas 2, seleccionar **Allow** com a efecte, i en **Principal** introduir `s3.amazonaws.com`. En **Action**, seleccionar **SendMessage**. En **Amazon Resource Name (ARN)**, introduir l'ARN de la cua de SQS (`arn:aws:sqs:us-east-1:707787736069:validatedCsvQueue`) i clicar a **Add Statement**.
     - En el pas 3, seleccionar **SendMessage** com a acció.
   - Clicar a **Generar política** i després a **Apply Policy**.

3. Configurar S3 per a enviar un missatge a la cua de SQS quan un fitxer és mogut a `validat/`:
   - Anar al bucket de S3 creat anteriorment i clicar a la pestanya **Propiedades**.
   - Desplaçar-se fins a **Notificaciones de eventos** i clicar a **Todos los eventos de creación de objetos**.
   - Proporcionar un nom per a la notificació (`Notificació fitxers validats`), seleccionar el prefix `validat/` i introduir el sufix `.csv`.
   - En **Tipo de evento**, seleccionar **All object create events**.
   - En **Destino**, seleccionar **Cola de SQS** i seleccionar la cua creada anteriorment (`validatedCsvQueue`).
   - Clicar a **Guardar cambios**.

4. Abans de crear la base de dades, hem de crear un grup de subxarxes de base de dades (DB subnet group) per a RDS, que és necessari per a desplegar la base de dades en una VPC. El grup de subxarxes defineix les subxarxes de la VPC on es poden desplegar les instàncies de base de dades. En aquest cas, seleccionarem les subxarxes privades creades anteriorment per a garantir que les instàncies de RDS no siguin accessibles des d'internet públic.
   - Anar al cercador de serveis i obrir **RDS**.
   - En el menú lateral, clicar a **Grupos de subredes** i després a **Crear grupo de subredes**.
   - Proporcionar un nom per al grup de subxarxes (`subxarxes-botigues`), en la descripció escriure una breu descripció (`Grup de subxarxes per a la base de dades de botigues`) i seleccionar la VPC creada anteriorment (`botigues-vpc`).

   - En la secció **Tipo de escalabilidad de clúster**, seleccionar **Aprovisionada**, en **Tipo de configuración aprovisionada** seleccionar **Clases ampliables (incluye clases t)** i **db.t3.small** com a tipus d'instància.

   - En l'apartat **Conectividad**, seleccionar la VPC creada anteriorment (`botigues-vpc`) i seleccione el grup de subxarxes creat anteriorment (`subxarxes-botigues`).

### Consulta dels fitxers CSV processats amb Athena (servei addicional)

En aquesta part s'utilitza Amazon Athena, com a servei addicional, per a consultar directament els fitxers CSV que han estat processats i emmagatzemats a la carpeta `processat/` del bucket de S3. D'aquesta manera es pot validar que les dades processades per Lambda es poden analitzar mitjançant consultes SQL sense necessitat de carregar-les prèviament a una base de dades relacional.

Athena permet crear una taula externa sobre els fitxers CSV ubicats a S3 i executar consultes SQL sobre aquests fitxers. Aquesta prova complementa la validació feta amb RDS, ja que permet comprovar les dades directament al data lake de S3.

1. Configurar la ubicació dels resultats de les consultes d'Athena:

* Anar al cercador de serveis i obrir **Athena**.
* Entrar a **Query editor**.
* Si és la primera vegada que s'utilitza Athena, configurar la ubicació on es desaran els resultats de les consultes.
* Indicar una ruta dins del bucket de S3, per exemple:

```text
s3://amzn-s3-uoc-2026-bucket/athena-results/
```

* Aquesta carpeta s'utilitzarà perquè Athena pugui escriure els resultats generats per les consultes SQL.

2. Crear una base de dades a Athena per al projecte:

* Des del Query editor d'Athena, executar la següent sentència SQL:

```sql
CREATE DATABASE IF NOT EXISTS botigues_athena;
```

* Un cop creada la base de dades, seleccionar-la al desplegable de l'esquerra o executar:

```sql
USE botigues_athena;
```

3. Crear una taula externa sobre els fitxers CSV processats:

Els fitxers CSV de vendes processades estan ubicats a la carpeta `processat/` del bucket de S3:

```text
s3://amzn-s3-uoc-2026-bucket/processat/
```

Per consultar-los amb Athena, es crea una taula externa que apunta directament a aquesta ubicació de S3:

```sql
CREATE EXTERNAL TABLE IF NOT EXISTS vendes_processades (
    id_botiga string,
    data string,
    hora string,
    id_producte string,
    departament string,
    talla string,
    preu double,
    unitats int,
    color string,
    descripcio string
)
ROW FORMAT SERDE 'org.apache.hadoop.hive.serde2.OpenCSVSerde'
WITH SERDEPROPERTIES (
    "separatorChar" = ",",
    "quoteChar" = "\"",
    "escapeChar" = "\\"
)
LOCATION 's3://amzn-s3-uoc-2026-bucket-ipc/processat/'
TBLPROPERTIES (
    "skip.header.line.count"="1"
);
```

Aquesta configuració indica a Athena que ha de llegir tots els fitxers CSV de la carpeta `processat/` com si fossin una única taula SQL. També s'indica que la primera línia dels fitxers correspon a la capçalera i no s'ha de tractar com una fila de dades.

4. Comprovar que Athena pot llegir les dades de S3:

* Executar una consulta de recompte per verificar que Athena detecta les files dels fitxers CSV processats:

```sql
SELECT COUNT(*) AS total_vendes_processades
FROM vendes_processades;
```

5. Validar les vendes més recents:

Per comprovar que les dades es poden ordenar i analitzar correctament, es pot executar una consulta semblant a la utilitzada amb RDS:

```sql
SELECT *
FROM vendes_processades
ORDER BY data DESC, hora DESC
LIMIT 5;
```

6. Fer comprovacions agregades sobre les dades processades:

Per comprovar les unitats venudes per producte, s'executa la consulta següent:

```sql
SELECT id_producte, SUM(unitats) AS total_unitats
FROM vendes_processades
GROUP BY id_producte
ORDER BY total_unitats DESC
LIMIT 10;
```

Per comprovar el volum de vendes per botiga:

```sql
SELECT id_botiga, COUNT(*) AS num_vendes, SUM(unitats) AS unitats_totals
FROM vendes_processades
GROUP BY id_botiga
ORDER BY num_vendes DESC;
```

Per calcular l'import total venut:

```sql
SELECT 
    SUM(preu * unitats) AS import_total
FROM vendes_processades;
```

## **Exercici 3**: KPIs Específics per a un Projecte d'Anàlisi de Dades a AWS *(2 punts)*

En un projecte d'anàlisi de dades desenvolupat a AWS (Amazon Web Services), els KPIs no només se centren en els resultats de la informació (la precisió de les dades o els insights generats), sinó també en el rendiment, l'eficiència i el cost de la infraestructura cloud.

Heu de fer un petit informe amb indicadors d'aquests 4 punts. Heu d'explicar quines accions proposaríeu per aconseguir els objectis que es detallen. A la descripció es detallen uns valors en % o en termes absoluts. Són simplement orientatius i pretenen il·lustrar com serien els KPIs en un cas real.


### 1. KPIs de Cost i Eficiència d'AWS (FinOps)

* **KPI:** Cost Mitjà per Càrrega de Treball
    * **Descripció:** El cost total d'AWS / Quantitat de dades processades o insights generats.
    * **Objectiu del Projecte:** Reduir el cost per unitat d'anàlisi en un 10%.
* **KPI:** Taxa d'Utilització dels Recursos
    * **Descripció:** Percentatge de temps en què els recursos d'AWS provisionats s'estan utilitzant activament.
    * **Objectiu del Projecte:** Mantenir per sobre del 80%.
* **KPI:** Cost d'Emmagatzematge per GB
    * **Descripció:** Cost mensual per mantenir les dades al data lake (S3 o equivalent).
    * **Objectiu del Projecte:** Optimitzar l'emmagatzematge amb polítiques de retenció adequades.

### 2. KPIs de Rendiment i Velocitat

* **KPI:** Temps de Latència de la Consulta
    * **Descripció:** Temps mitjà de resposta de consultes crítiques (en segons).
    * **Objectiu del Projecte:** Reduir la latència de les consultes clau per sota de 5 segons.
* **KPI:** Temps d'Execució del Pipeline ETL
    * **Descripció:** Temps total (en minuts) des de l'ingesta de dades fins a la disponibilitat per a l'anàlisi (AWS Glue/Step Functions).
    * **Objectiu del Projecte:** Reduir el temps d'execució del pipeline ETL en un 15%.
* **KPI:** Taxa de Creixement de Dades
    * **Descripció:** GB/TB de dades noves ingerit al dia/mes.
    * **Objectiu del Projecte:** Assegurar que el sistema (Kinesis/MSK) pugui escalar per gestionar aquest volum.

### 3. KPIs de Qualitat i Fiabilitat de les Dades

* **KPI:** Taxa d'Error d'Ingesta
    * **Descripció:** Percentatge de registres o fitxers que fallen en la validació durant la càrrega a AWS S3.
    * **Objectiu del Projecte:** Mantenir la taxa d'error d'ingesta per sota del 0,5%.
* **KPI:** Percentatge d'Inconsistència de Dades
    * **Descripció:** Número d'entrades no vàlides / Total d'entrades analitzades.
    * **Objectiu del Projecte:** Eliminar o minimitzar les dades inconsistents.
* **KPI:** Disponibilitat del Servei
    * **Descripció:** Temps que els serveis d'anàlisi (Redshift, QuickSight) estan operatius (en percentatge).
    * **Objectiu del Projecte:** Mantenir la disponibilitat del servei per sobre del 99,9%.

### 4. KPIs d'Impacte i Adopció

* **KPI:** Adopció dels Informes
    * **Descripció:** Usuaris actius diaris/setmanals que accedeixen als dashboards (QuickSight).
    * **Objectiu del Projecte:** Augmentar l'ús dels dashboards clau en un 20% en tres mesos.
* **KPI:** Generació d'Ingressos (o Estalvi) Atribuïble
    * **Descripció:** Valor monetari generat o estalviat gràcies als insights del projecte.
    * **Objectiu del Projecte:** Demostrar un ROI positiu per al projecte.
* **KPI:** Precisió del Model Predictiu (si n'hi ha)
    * **Descripció:** Mètrica clau del model de Machine Learning (Puntuació F1, AUC, o Precisió).
    * **Objectiu del Projecte:** Assolir una precisió del model del 90% o superior.

#### 2.1 Script de Generació de Fitxers CSV de Prova

Aquest script en Python genera fitxers CSV amb dades de vendes simulades per a
les botigues. Utilitza la biblioteca csv per escriure els fitxers i random per
generar dades aleatòries. Els fitxers es desen localment a la carpeta `data` i
després es poden pujar a S3 a la carpeta `entrada`.

L'script genera una base dades de productes i després crea fitxers CSV amb
registres de vendes per a 30 botigues. Si es torna executar l'script es poden
generar fitxers amb dades de vendes diferents reutilitzant els productes.

El fitxer productes.csv té la següent estructura:

- id_producte: identificador únic del producte (cadena alfanumèrica de 12
  caràcters)
- departament: categoria a la qual pertany el producte (Ex. home, dona, nen o
  nena)
- descripcio: breu descripció del producte
- preu: preu unitari del producte

Cada fitxer de vendes té la següent estructura:

- id_botiga: identificador de la botiga (Ex. BAR01, BAR02, LLE01, etc.)
- data: data de la venda en format YYYY-MM-DD
- hora: hora de la venda en format HH:MM:SS
- id_producte: identificador del producte venut (ha de coincidir amb un
  `id_producte` del fitxer de productes)
- departament: categoria del producte venut (ha de coincidir amb el departament
  del fitxer de productes)
- talla: talla del producte venut (Ex. S, M, L, XL, etc.)
- preu: preu unitari del producte venut (ha de coincidir amb el preu del fitxer
  de productes)
- unitats: nombre de unitats venudes
- descripcio: breu descripció de la venda (ha de coincidir amb la descripció del
  fitxer de productes)

#### 2.2 Script de Lambda per a la validació dels fitxers CSV

Aquest script és una funció Lambda en Python que s’encarrega de validar els
fitxers CSV que arriben a la carpeta `entrada/` de S3. La funció es dispara amb
un esdeveniment de S3 quan es puja un nou fitxer.

El codi llegeix el fitxer CSV, comprova que té la capçalera correcta i que cada
registre compleix amb diverses validacions.

Si el fitxer és vàlid, es mou a la carpeta `validat/`. Si hi ha algun error, es
mou a la carpeta `error/` i es registra l’error a CloudWatch Logs.

El codi de la funció Lambda utilitza la biblioteca boto3 per interactuar amb S3
i la biblioteca csv per llegir els fitxers CSV.

#### 2.3 Script de Lambda per al processament i càrrega a RDS

Aquest script és una funció Lambda en Python que processa els fitxers CSV que
han passat la validació i que es troben a la carpeta `validat/` de S3. Aquesta
funció es dispara amb un esdeveniment de SQS, ja que quan s'afegeix un fitxer a
la carpeta `validat/`, s'envia un missatge a una cua SQS que activa aquesta
funció Lambda.

El codi llegeix el fitxer CSV, processa les dades i les carrega a la base de
dades RDS compatible amb MySQL. Utilitza la biblioteca `pymysql` per connectar-se
a la base de dades i executar les insercions. El codi també gestiona errors de
connexió i d'inserció, a més es manté un registre dels resultats parcials a
CloudWatch.

La biblioteca `pymysql` no està disponible per defecte a Lambda, per això s’ha
creat una capa de Lambda que inclou aquesta biblioteca i s'ha adjuntat a la
funció.

La biblioteca `boto3` també s'utilitza per llegir els fitxers de S3 i moure'ls
després de processar-los correctament. Si el processament és correcte, el fitxer
es mou a la carpeta `processat/`.

Per a aquest script, és important configurar correctament les variables d'entorn
per a la connexió a RDS, com ara el nom d'usuari, contrasenya, nom de la base de
dades i l'endpoint.

entre les taules. El codi SQL es pot executar des d’una instància EC2 amb un
